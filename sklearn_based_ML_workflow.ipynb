1.Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
```

2. Creating Random Datasets

```python
X=np.random.rand(20).reshape(-1,1) #1 means 2D array me ab rows according to situation ajust krdo...jaise yaha 1 column chahiye toh 20 rows banengi
print(X)
Y=np.random.rand(20)
Y
```

3. Splitting data into train and test dataset

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.3,random_state=20)
print("X_train",X_train)
print("Y_train",Y_train)
print("X_test",X_test)
print("Y_test",Y_test)
```

4. Creating an instance for the linear regression model

```python
lrModel=LinearRegression()
lrModel
```

5. Creating/Train Linear Model

```python
lrModel.fit(X_train,Y_train)
lrModel
```

6. Make prediction on test dataset

```python
Y_pred=lrModel.predict(X_test)
Y_pred
```

7. Model Evaluation

7.1 Get the coefficients

```python
lrModel.coef_   #tells Weight
```

```python
lrModel.intercept_ #tells bias
```

```python
r2_score(Y_test,Y_pred)
```

```python
mean_squared_error(Y_test,Y_pred)
```

Scatter Plot

```python
# Plotting the output
plt.scatter(X_test, Y_test, color = "black", label = "original data")
plt.plot(X_test, Y_pred, color = "blue", linewidth=3, label = "regression line")
plt.xlabel("Independent Feature")
plt.ylabel("Target Values")
plt.title("Simple Linear Regression")
#plt.show()
```
