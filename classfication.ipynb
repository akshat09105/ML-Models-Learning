3. Classification: Apply "Data Preprocessing"

3.1 Model performance using "Normalization"

```python
setup(data=mydataset,target="Class variable",normalize=True,normalize_method="zscore")
cm=compare_models()
```

3.2 Model performance using "Feature Selection"

```python
setup(data=mydataset,target="Class variable",feature_selection=True,feature_selection_method="univariate")
cm=compare_models()
```

3.3 Model performance using "Outlier Removal"

```python
setup(data=mydataset,target="Class variable",outliers_method="0.5",verbose=True)
cm=compare_models()
```

3.4 Model performance using "Transformation"

```python
setup(data=mydataset,target="Class variable",transformation=True,transformation_method="yeo-johnson")
cm=compare_models()
```

3.5 Model performance using "PCA"

```python
setup(data=mydataset,target="Class variable",pca=True,pca_method="linear",verbose=False)
cm=compare_models()
```

3.6 Model performance using "Outlier Removal" + "Normalization"

```python
s=setup(data=mydataset,target="Class variable",outliers_method="0.5",normalize=True,normalize_method="zscore",verbose=False)
cm=compare_models()
```

3.7 Model performance using "Outlier Removal" + "Normalization" + "Transformation"

```python
s=setup(data=mydataset,target="Class variable",outliers_method="0.5",normalize=True,normalize_method="zscore",transformation=True,transformation_method="yeo-johnson",verbose=False)
cm=compare_models()
```
